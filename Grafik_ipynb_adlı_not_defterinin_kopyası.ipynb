```python
import pandas as pd
import matplotlib.pyplot as plt

# Veri setini yükle
data = pd.read_csv('football_mini.csv')
oyucuYasi = data['Age']
milliyet = data['Nationality']


# Dağılım grafiği oluştur
plt.figure(figsize=(10, 6))
plt.bar(oyucuYasi, milliyet)
plt.title('futbol oyuncularınınn milliyetlerine göre yaş dağılımı')
plt.xlabel('Yaş')
plt.ylabel('milliyet')
plt.show()
```
